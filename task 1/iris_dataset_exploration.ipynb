# Task 1: Exploring and Visualizing the Iris Dataset

## Objective
Learn how to load, inspect, and visualize a dataset to understand data trends and distributions.

## Dataset
The Iris dataset is a classic dataset in machine learning and statistics. It contains measurements of 150 iris flowers from three different species: Setosa, Versicolor, and Virginica.

## Skills Covered
- Data loading and inspection using pandas
- Descriptive statistics and data exploration
- Basic plotting and visualization with seaborn and matplotlib

## 1. Import Required Libraries
Import pandas for data handling, and matplotlib.pyplot and seaborn for visualization.

```python
# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style for better visualizations
plt.style.use('default')
sns.set_palette("husl")

# Display all columns and rows without truncation
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print("Libraries imported successfully!")
```

## 2. Load the Iris Dataset
Load the Iris dataset into a pandas DataFrame using seaborn's built-in dataset.

```python
# Load the Iris dataset using seaborn
iris_df = sns.load_dataset('iris')

print("Dataset loaded successfully!")
print(f"Dataset type: {type(iris_df)}")

# Alternative method: Loading from URL if seaborn doesn't work
# iris_url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
# iris_df = pd.read_csv(iris_url)
```

## 3. Inspect Dataset Structure
Print the shape of the DataFrame, column names, and display the first few rows using .head().

```python
# Dataset Shape
print("Dataset Shape:")
print(f"Rows: {iris_df.shape[0]}, Columns: {iris_df.shape[1]}")
print()

# Column Names
print("Column Names:")
print(iris_df.columns.tolist())
print()

# First few rows
print("First 5 rows of the dataset:")
print(iris_df.head())
print()

# Last few rows
print("Last 5 rows of the dataset:")
print(iris_df.tail())
```

## 4. Summary Statistics
Use .info() and .describe() to display summary statistics and data types.

```python
# Dataset Info
print("Dataset Information:")
print(iris_df.info())
print()

# Descriptive Statistics
print("Descriptive Statistics:")
print(iris_df.describe())
print()

# Check for missing values
print("Missing Values:")
print(iris_df.isnull().sum())
print()

# Unique values in the species column
print("Unique Species:")
print(iris_df['species'].unique())
print()

# Count of each species
print("Species Count:")
print(iris_df['species'].value_counts())
```

## 5. Scatter Plot of Feature Relationships
Create scatter plots to visualize relationships between different features using seaborn's pairplot and scatterplot.

```python
# Create pairplot to show relationships between all features
plt.figure(figsize=(12, 10))
sns.pairplot(iris_df, hue='species', height=2.5)
plt.suptitle('Pairplot of Iris Dataset Features', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()

# Individual scatter plots for specific feature pairs
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Sepal Length vs Sepal Width
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', 
                hue='species', s=100, alpha=0.8, ax=axes[0,0])
axes[0,0].set_title('Sepal Length vs Sepal Width')

# Sepal Length vs Petal Length
sns.scatterplot(data=iris_df, x='sepal_length', y='petal_length', 
                hue='species', s=100, alpha=0.8, ax=axes[0,1])
axes[0,1].set_title('Sepal Length vs Petal Length')

# Petal Length vs Petal Width
sns.scatterplot(data=iris_df, x='petal_length', y='petal_width', 
                hue='species', s=100, alpha=0.8, ax=axes[1,0])
axes[1,0].set_title('Petal Length vs Petal Width')

# Sepal Width vs Petal Width
sns.scatterplot(data=iris_df, x='sepal_width', y='petal_width', 
                hue='species', s=100, alpha=0.8, ax=axes[1,1])
axes[1,1].set_title('Sepal Width vs Petal Width')

plt.tight_layout()
plt.show()
```

## 6. Histograms of Feature Distributions
Plot histograms for each feature to show their value distributions using matplotlib and seaborn.

```python
# Create histograms for all numerical features
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# List of numerical columns
numerical_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Create histograms for each feature
for i, column in enumerate(numerical_columns):
    row = i // 2
    col = i % 2
    
    # Create histogram with different colors for each species
    for species in iris_df['species'].unique():
        species_data = iris_df[iris_df['species'] == species][column]
        axes[row, col].hist(species_data, alpha=0.7, label=species, bins=15)
    
    axes[row, col].set_title(f'Distribution of {column.replace("_", " ").title()}')
    axes[row, col].set_xlabel(column.replace("_", " ").title())
    axes[row, col].set_ylabel('Frequency')
    axes[row, col].legend()
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Alternative: Using seaborn's histplot
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for i, column in enumerate(numerical_columns):
    row = i // 2
    col = i % 2
    
    sns.histplot(data=iris_df, x=column, hue='species', 
                 multiple="stack", alpha=0.8, ax=axes[row, col])
    axes[row, col].set_title(f'Distribution of {column.replace("_", " ").title()}')

plt.tight_layout()
plt.show()
```

## 7. Box Plots for Outlier Detection
Create box plots for each feature to identify potential outliers using seaborn's boxplot.

```python
# Create box plots for all numerical features
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

numerical_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, column in enumerate(numerical_columns):
    row = i // 2
    col = i % 2
    
    sns.boxplot(data=iris_df, x='species', y=column, ax=axes[row, col])
    axes[row, col].set_title(f'Box Plot of {column.replace("_", " ").title()} by Species')
    axes[row, col].set_xlabel('Species')
    axes[row, col].set_ylabel(column.replace("_", " ").title())

plt.tight_layout()
plt.show()

# Overall box plots without species grouping
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

for i, column in enumerate(numerical_columns):
    sns.boxplot(data=iris_df, y=column, ax=axes[i])
    axes[i].set_title(f'Box Plot of {column.replace("_", " ").title()}')
    axes[i].set_ylabel(column.replace("_", " ").title())

plt.tight_layout()
plt.show()

# Identify potential outliers
print("Potential Outliers Analysis:")
for column in numerical_columns:
    Q1 = iris_df[column].quantile(0.25)
    Q3 = iris_df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = iris_df[(iris_df[column] < lower_bound) | (iris_df[column] > upper_bound)]
    
    print(f"\n{column.replace('_', ' ').title()}:")
    print(f"  Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"  Lower bound: {lower_bound:.2f}, Upper bound: {upper_bound:.2f}")
    print(f"  Number of outliers: {len(outliers)}")
    
    if len(outliers) > 0:
        print(f"  Outlier values: {outliers[column].tolist()}")
    else:
        print("  No outliers detected")
```
